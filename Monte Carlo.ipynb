# Monte Carlo algorithm

A Monte Carlo algorithm is an algorithm for computers which is used to simulate the behaviour of other systems. It is not an exact method, but a heuristical one, typically using randomness and statistics to get a result. The algorithm terminates with an answer that is correct with probability p<1.

It is a computation process that uses random numbers to produce an outcome(s). Instead of having fixed inputs, probability distributions are assigned to some or all of the inputs. This will generate a probability distribution for the output after the simulation is run.

## Estimation of Pi Value

The goal is to simulate random (x, y) points in a 2-D plane with a domain of 1 unit square. Imagine a circle with the same diameter as the square inside the same domain. The ratio of the number of points inside the circle to the total number of generated points is then calculated to estimate the  $\pi$ value.

<img src="pi.png">

As we know that area of the square is 1 unit square while that of circle is $\pi * (\frac{1}{2})^{2} = \frac{\pi}{4}$  . Now for a very large number of generated points,

$$\frac{\text{area of the circle}}{\text{area of the square}} = \frac{\text{no. of points generated inside the circle}}{\text{no. of points generated inside the square}}$$

Therefore, $$\pi = 4 * \frac{\text{no. of points generated inside the circle}}{\text{no. of points generated inside the square}}$$


We just create random (x, y) pairs and check whether $x^{2} + y^{2} \leqslant 1$. If this is the case, the number of points that appear inside the circle is increased. The higher the number of iterations in randomization and simulated algorithms like Monte Carlo, the more accurate the output.

```python
import random
sim_num= 1000
circle_points= 0
square_points= 0

# total Random numbers generated= possible x values* possible y values
for i in range(sim_num**2):
    
    # Randomly generated x and y values from a uniform distribution
    # Range of x and y values is -1 to 1
    rand_x= random.uniform(-1, 1)
    rand_y= random.uniform(-1, 1)
  
    # distance between (x, y) from the origin
    origin_dist= rand_x**2 + rand_y**2
  
    # checking if (x, y) lies inside the circle
    if origin_dist<= 1:
        circle_points+= 1
  
    square_points+= 1
  
    pi = 4* circle_points/ square_points

print(" The estimation of Pi=", pi)
```
