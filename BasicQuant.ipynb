## Retriving data using yfinance

```python
import yfinance as yf
data = yf.Ticker("IBM")
data.info
```

#### Business summary of selected company

```python
data.info['longBusinessSummary']
```

#### Download historical data

```python
df = data.history(period = "max",rounding=2)
df.head(5)
```

####  Download call and put data information

```python
# List of all options Maturity Date 
data.options
```

#####  a. Call Option data at specific data

```python
opt = data.option_chain(date='2022-02-04')
opt.calls
```

##### Puts data

```python
opt.puts
```
